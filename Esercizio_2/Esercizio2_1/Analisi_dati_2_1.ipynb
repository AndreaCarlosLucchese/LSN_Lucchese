```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
plt.rcParams.update({'font.size': 14})
```

```python
sum_prog= np.loadtxt("Int_media.txt")
sum_prog2= np.loadtxt("Int_samp.txt")
```

In questo esercizio mostro due diversi metodi di integrazione per il calcolo del seguente integrale:
 $I = \int_0^1 \frac{\pi}{2}\cos(\pi x/2) dx$.
Entrambi sono metodi di integrazione Monte Carlo, nel primo caso utlizzo il metodo della media.
Nel secondo caso uso invece il metodo dell' "importance sampling". La distribuzione scelta per ottenere l'importance sampling è un retta di equazione : $ y=x-1$. Per estrarre numeri casuali da questa funzione è stato il utilizzato il metodo dell "Inversione della Cumulata".

```python
x = np.arange(100)
plt.figure(figsize=(20,10));
plt.errorbar(x,sum_prog[:,0]-1,yerr=sum_prog[:,1])
plt.xlabel('#Blocchi')
plt.ylabel('<r>-1/2')
plt.title("Integrale Media")
plt.grid(True)
plt.show()
```

```python
x = np.arange(100)
plt.figure(figsize=(20,10));
plt.errorbar(x,sum_prog2[:,0]-1,yerr=sum_prog2[:,1])
plt.xlabel('#Blocchi')
plt.ylabel('<r>-1/2')
plt.title("Integrale importance sampling")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(20,10));
plt.plot(x,(sum_prog[:,1]),label="Integrale Media")
plt.plot(x,(sum_prog2[:,1]),label="Integrale importance sampling")
plt.xlabel('#Blocchi')
plt.ylabel('Err')
plt.title("Confronto tra gli errori dei due metodi di integrazione")
plt.grid(True)
plt.legend()
plt.show
```

Nel grafico sopra si può notare come utilizzando l'importance sampling l'errore Monte Carlo nel calcolo dell'integrale diminuisce sensibilmente.
