```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
plt.rcParams.update({'font.size': 14})
```

```python
sum_prog= np.loadtxt("RW_discreto.txt")
sum_prog1= np.loadtxt("RW_continuo.txt")
```

In questo esercizio ho costruito due random walk, uno discreto ed uno continuo. Il primo si muove di un passo $a=1$ in un reticolo cubico. Il secondo invece si muove in uno spazio continuo sempre con passo $a=1$. 

```python
x = np.arange(100)
plt.figure(figsize=(20,10));
plt.errorbar(x,sum_prog[:,0],yerr=sum_prog[:,1],label="Random Walk")
plt.plot(np.sqrt(x),label= "Fit")
plt.xlabel('#Passi')
plt.ylabel('Distanza')
plt.title("Random Walk discreto")
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```

```python
x = np.arange(100)
plt.figure(figsize=(20,10));
plt.errorbar(x,sum_prog1[:,0],yerr=sum_prog1[:,1],label="Random Walk")
plt.plot(np.sqrt(x),label= "Fit")
plt.xlabel('#Passi')
plt.ylabel('Distanza')
plt.title("Random Walk continuo")
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```

In entrambi ho fittato i dati della mia simulazione con la funzione $f(N)=k\sqrt{N}$, con $N$ numero di passi. Per in entrambi i random walk la funzione $f(N)$ fitta molto bene i dati, perci possiamo affermare che entrambi sono processi diffusivi.
