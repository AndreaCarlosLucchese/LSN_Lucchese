```python
import numpy as np                                                                 
import matplotlib.pyplot as plt    
```

In questo esercizio ho sviluppato un codice che sfrutta un algoritmo genetico per risolvere il problema del "Travelling Salesman".

Ho costruito 3 classi diverse. 

La prima classe chiamata *Cities*, gestiva la posizione delle diverse città e permettava il calcolo della distanza tra due città e quindi il calcolo della funzione costo L1 (nel mio codice ho implementato solo L1).

La seconda classe chiamata *Salesman_Path*, gestiva un singolo possibile percorso del venditore. Perciò inizializzava un percorso casuale e ne calcolava la funzione di costo L1 totale. All'interno di questa classe ho sviluppato pure le mutazioni che possono avvenire nella generazione, ho scelto di utilizzare: Singola permutazione, Shift e inversione.

La terza classe chamata *Genetic_Algo*, gestiva l'intera popolazione ordinandola in funzione di L1. Inoltre aveva l'algortimo di Cross-Over che aveva una probabilità del 60% di avvenire, la scelta dei due elementi della popolazione è stata fatta utilizzando la seguente formula :$j=N\times int(r^5)$

Ho scelto di utilizzare una popolazione di 400 individui e ho implementato anche un piccolo algoritmo di convergenza per capire quando fossi arrivato all'L1 ottimale.
 


## CIRCLE 32 CITIES


```python
x=np.loadtxt("outputs/city_circle.dat")                            
fig,ax = plt.subplots(1,3, figsize=(20, 6))

ax[0].plot(x[:,0], x[:,1], marker=".", linestyle="");
ax[0].set_title("Cities locations");

initial=np.loadtxt("outputs/circle_initial_best_path.dat",dtype="uint32")  
x_initial=x[initial,0]
y_initial=x[initial,1]
ax[1].plot(x_initial, y_initial, marker="."); 
ax[1].set_title("Initial best path");

final=np.loadtxt("outputs/circle_new_best_path.dat",dtype="uint32")  
x_final=x[final,0]
x_final=np.append(x_final, x[final[0],0])
y_final=x[final,1]
y_final=np.append(y_final, x[final[0],1])
ax[2].plot(x_final, y_final, marker=".");  
ax[2].set_title("Final L1 best path");
```

```python
av=np.loadtxt("outputs/av_circle.dat") 
best=np.loadtxt("outputs/best_circle.dat")

plt.figure(figsize=(20,10));

plt.plot(av, label="AV")
plt.plot(best,label="Best")
plt.title("L1")
plt.legend()
```
