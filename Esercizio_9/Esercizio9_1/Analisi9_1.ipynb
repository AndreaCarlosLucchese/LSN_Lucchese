```python
import numpy as np                                                                 
import matplotlib.pyplot as plt    
```

Il mio algoritmo genetico consiste in un crossover che può avvenire con una probabilità del 60% per generare una nuova popolazione. Sulla nuova popolazione viene attuato una mutazione con probabilità 10%. Inoltre ho scritto un piccolo algoritmo di convergenza che verificasse la convergenza ogni 100 generazioni andando a guardare la differenza tra le funzioni costo.

## CIRCLE 32 CITIES


```python
x=np.loadtxt("outputs/city_circle.dat")                            
fig,ax = plt.subplots(1,3, figsize=(20, 6))

ax[0].plot(x[:,0], x[:,1], marker=".", linestyle="");
ax[0].set_title("Cities locations");

initial=np.loadtxt("outputs/circle_initial_best_path.dat",dtype="uint32")  
x_initial=x[initial,0]
y_initial=x[initial,1]
ax[1].plot(x_initial, y_initial, marker="."); 
ax[1].set_title("Initial best path");

final=np.loadtxt("outputs/circle_new_best_path.dat",dtype="uint32")  
x_final=x[final,0]
x_final=np.append(x_final, x[final[0],0])
y_final=x[final,1]
y_final=np.append(y_final, x[final[0],1])
ax[2].plot(x_final, y_final, marker=".");  
ax[2].set_title("Final L1 best path");
```

```python
av=np.loadtxt("outputs/av_circle.dat") 
best=np.loadtxt("outputs/best_circle.dat")

plt.figure(figsize=(20,10));

plt.plot(av, label="AV")
plt.plot(best,label="Best")
plt.title("L1")
plt.legend()
```
