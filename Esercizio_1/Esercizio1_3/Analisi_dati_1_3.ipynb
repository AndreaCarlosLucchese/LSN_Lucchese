```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
plt.rcParams.update({'font.size': 20})
```

```python
sum_prog= np.loadtxt("result_pi.txt")
```

Nel grafico mostro il risultato della simulazione Monte Carlo per il calcolo del valore di $\pi$ con l'esperimento di Buffon. Per calcolare il valore dell'angolo $\theta$ a cui cadeva il bastoncino ho utilizzato il metodo dell'accept reject.

```python
x = np.arange(100)
plt.figure(figsize=(20,10));
plt.errorbar(x,sum_prog[:,0]-np.pi,yerr=sum_prog[:,1],label="Sim")
plt.hlines(0,-1,101,color="r",label="Teo")
plt.xlabel('#Blocchi',)
plt.ylabel('$\pi_M$-$\pi$')
plt.grid(True)
plt.legend()
plt.show()
```

Nel grafico sulle ordinate è rappresentato la differenza tra il valore calcolato della mia simulazione Monte Carlo ed il valore teorico. Sulle ascisse il numero di blocchi utilizzato per fare la media a blocchi. Mentre le barre di errore sono quelle ottenute dalla media blocchi.
