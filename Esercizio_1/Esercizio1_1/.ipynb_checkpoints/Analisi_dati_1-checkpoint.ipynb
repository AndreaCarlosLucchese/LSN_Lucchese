## MEDIA

Copio i file ottenuti dalle simulazioni in C++ e costruisco il grafico che mostra l'andamento della media in funzione del numero di lanci. Le barre di errore di ogni punto rapprensentano l'errore della media a blocchi. Aumentando il numero di lanci per blocco la media diventa pi√π precisa e l'errore che facciamo nel suo calcolo diminuisce.

```python
M=100000
N=100
L=int(M/N) 
x = np.arange(N)
x *= L # Number of throws = block * (Number of throws in each block)
```

```python
plt.figure(figsize=(20,10));

sum_prog= np.loadtxt("sum_prog.txt")
err= np.loadtxt("Err.txt")

plt.errorbar(x,sum_prog-0.5,yerr=err)
plt.xlabel('#throws')
plt.ylabel('<r>-1/2')
plt.grid(True)
plt.show()
```

## Varianza

Eseguo la stessa operazione utilizzata per la media per ottenere il grafico della varianza in funzione del numero di lanci. L'andamento del grafico segue lo stesso della media.

```python
plt.figure(figsize=(20,10));

sum_prog_var= np.loadtxt("sum_prog_var.txt")
err_var= np.loadtxt("Err_var.txt")
plt.errorbar(x,sum_prog_var-1/12,yerr=err_var)
plt.xlabel('#throws')
plt.ylabel('<(r-0.5)^2>-1/12')
plt.grid(True)
plt.show()
```
