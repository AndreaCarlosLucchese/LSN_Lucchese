```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
plt.rcParams.update({'font.size': 16})
```

```python
sum_prog_m= np.loadtxt("ISING_1D/energy_graph_m.txt")
sum_prog_g= np.loadtxt("ISING_1D/energy_graph_g.txt")
heat_m=np.loadtxt("ISING_1D/cap_graph_m.txt")
heat_g=np.loadtxt("ISING_1D/cap_graph_g.txt")
chi_m=np.loadtxt("ISING_1D/chi_graph_m.txt")
chi_g=np.loadtxt("ISING_1D/chi_graph_g.txt")
mag_m=np.loadtxt("ISING_1D/mag_graph_m.txt")
mag_g=np.loadtxt("ISING_1D/mag_graph_g.txt")

points=100
T = np.linspace(0.5,2.0,num=points)
beta = 1/T
J = 1.0
Ns = 50
th = np.tanh(J/T)
thN= th**Ns
ch = 1/th
e = -J*( th + ch*thN )/( 1 + thN )
```

In questo esercizio calcolo l'energia, il calore specifico, la magnetizzazione e la capacità di un modello di Ising $50$ spins and $J=1$. Per farlo ho campionato la distribuzione di Boltzmann , definita come $p(\{ s_i \}) = \exp(-\beta H)/Z$, utilizzando l'algoritmo di Metropolis e l'algoritmo di Gibbs. Nell'algoritmo di Gibbs ogni mossa viene accettata e utilizzo la probabilità condizionata. Per campionare la distribuzione ad ogni mossa giravo uno spin casuale e valutavo se accettare o meno la mossa.

Ho campionato le 4 diverse quantità a 20 diverse temperature dell'intervallo [0.5,2]. Per equilibrare il sistema ho aggiunto la possibilità di ripartire da una determinata configurazione degli spini, perciò ho fatto una prima simulazione con una configurazione casuale degli spin e ne ho salvato la posizione finale. Sucessivamente ho cominciato con le misurazioni. 

Ho eseguito un totale di $M=5x10^5$ simulazioni per ogni singola temperatura, ed ho utilizzato una media a blocchi con $100$ blocchi.

```python
plt.figure(figsize=(20,10));
plt.errorbar(sum_prog_m[:,0],sum_prog_m[:,1],yerr=sum_prog_m[:,2],label='Metropolis')
plt.errorbar(sum_prog_g[:,0],sum_prog_g[:,1],yerr=sum_prog_g[:,2],label='Gibbs')
plt.plot(T, e, label="Teorico")
plt.legend()
plt.grid(True)
plt.title('Ising 1D, Energia interna')
plt.xlabel("Temperatura")
plt.ylabel("Energia/N" )
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.errorbar(heat_m[:,0],heat_m[:,1],yerr=heat_m[:,2],label='Metropolis')
plt.errorbar(heat_g[:,0],heat_g[:,1],yerr=heat_g[:,2],label='Gibbs')
heat=((beta*J)**2)*(((1+thN+(Ns-1)*(th**2)+(Ns-1)*(ch**2)*thN)/(1+thN))-Ns*((th+ch*thN)/(1+thN))**2)
plt.plot(T, heat, label ="Teorico")
plt.legend()
plt.grid(True)
plt.title('Ising 1D, Capacità')
plt.xlabel("Temperatura")
plt.ylabel("Capacità/N" )
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.errorbar(chi_m[:,0],chi_m[:,1],yerr=chi_m[:,2],label='Metropolis')
plt.errorbar(chi_g[:,0],chi_g[:,1],yerr=chi_g[:,2],label='Gibbs')
X = beta*np.exp(2*beta*J)*(1-thN)/(1+thN)
plt.plot(T, X, label="Teorico")
plt.legend()
plt.grid(True)
plt.title('Ising 1D, Suscettività')
plt.xlabel("Temperatura")
plt.ylabel("Suscettività/N" )
plt.show()

```

```python
plt.figure(figsize=(20,10))
plt.errorbar(mag_m[:,0],mag_m[:,1],yerr=mag_m[:,2],label='Metropolis')
plt.errorbar(mag_g[:,0],mag_g[:,1],yerr=mag_g[:,2],label='Gibbs')
h=0.02 #external field
b = 1/T

l1 = np.exp(b*J)*np.cosh(b*h)+np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
l2 = np.exp(b*J)*np.cosh(b*h)-np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
Z = l1**Ns + l2**Ns
M = (np.exp(b*J)*np.sinh(b*h)*((l1**(Ns-1))*(1+np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))) 
        + (l2**(Ns-1))*(1-np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J)))))/(Z)
plt.plot(T, M, label="Teorico")
plt.legend()
plt.grid(True)
plt.title('Ising 1D, Magnetizzazione')
plt.xlabel("Temperature")
plt.ylabel("Magnetizzazione/N" )
plt.show()
```

Nei grafici sopra mostro i risulati delle mie simulazioni utilizzando l'algoritmo di Metropolis e l'algoritmo di Gibbs. Il codice stima molto bene l'energia interna, la suscettività e la magnetizzazione. Il codice ha difficoltà,invece, a stimare la capacità a basse temperatura, sotto stimando il valore.
