# Analisi esercizio 5.1

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
plt.rcParams.update({'font.size': 17})
```

```python
sum_prog= np.loadtxt("result_100_unif.txt")
sum_prog2= np.loadtxt("result_100_gauss.txt")
sum_prog3= np.loadtxt("result_210_unif.txt")
sum_prog4= np.loadtxt("result_210_gauss.txt")
```

In questo esercizio abbiamo utilizzato l'alogoritmo di Metropolis con 2 diverse distribuzioni, una distribuzione uniforme ed una gaussiana, con $\mu=0$ e $\sigma=1$, per campionare il raggio medio di un atomo di idrogenoide. Quest'ultimo Ã¨ stato stimato per lo stato fondamentale e per il primo stato eccitato dell'idrogenoide.

```python
x = np.arange(100)
plt.figure(figsize=(20,10));
plt.errorbar(x,sum_prog[:,0],yerr=sum_prog[:,1],label="Distribuzione Uniforme")
plt.errorbar(x,sum_prog2[:,0],yerr=sum_prog2[:,1],label="Distribuzione Gaussiana ")
plt.xlabel('#Blocchi')
plt.ylabel('Raggio')
plt.title("Stato Fondamentale")
plt.legend()
plt.grid(True)
plt.show()
```

```python
x = np.arange(100)
plt.figure(figsize=(20,10));
plt.errorbar(x,sum_prog3[:,0],yerr=sum_prog3[:,1],label="Distribuzione Uniforme")
plt.errorbar(x,sum_prog4[:,0],yerr=sum_prog4[:,1],label="Distribuzione Gaussiana ")
plt.xlabel('#Blocchi')
plt.ylabel('Raggio')
plt.title("Primo stato eccitato")
plt.legend()
plt.grid(True)
plt.show()
```
