```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
sum_prog= np.loadtxt("result_100_unif.txt")
sum_prog2= np.loadtxt("resul_100_gauss.txt")
sum_prog3= np.loadtxt("result_210_unif.txt")
sum_prog4= np.loadtxt("resul_210_gauss.txt")
```
