```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
plt.rcParams.update({'font.size': 18})
```

```python
#parameters for argon
sigma = 0.34    
epsilon_boltzman = 120  #e/kB
k_boltzman=1.38064852*1E-23
epsilon=epsilon_boltzman*k_boltzman
sigma_si=sigma*1E-9
mass = 39.948*1.66054*1E-27
```

## Argon Liquido

```python
ave_kin= np.loadtxt("ave_ekin.out")
ave_pot= np.loadtxt("ave_epot.out")
ave_tot= np.loadtxt("ave_etot.out")
ave_temp= np.loadtxt("ave_temp.out")
```

```python
x = np.arange(100)
fig, ax = plt.subplots(2,2,figsize=(20, 15))

ax[0,0].errorbar(x,ave_kin[:,0]*epsilon,yerr=ave_kin[:,1]*epsilon)
ax[0,1].errorbar(x,ave_pot[:,0]*epsilon,yerr=ave_pot[:,1]*epsilon)
ax[1,0].errorbar(x,ave_tot[:,0]*epsilon,yerr=ave_tot[:,1]*epsilon)
ax[1,1].errorbar(x,ave_temp[:,0]*epsilon_boltzman,yerr=ave_temp[:,1]*epsilon_boltzman)
ax[0,0].set_xlabel("# Blocchi")
ax[0,1].set_xlabel("# Blocchi")
ax[1,0].set_xlabel("# Blocchi")
ax[1,1].set_xlabel("# Blocchi")

ax[0,0].set_title('Energia Cinetica')
ax[0,1].set_title('Energia Potenziale')
ax[1,0].set_title('Energia Totale')
ax[1,1].set_title('Temperatura')

plt.show
```
