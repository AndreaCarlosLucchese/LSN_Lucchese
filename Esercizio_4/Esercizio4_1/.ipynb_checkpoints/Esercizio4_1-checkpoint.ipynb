```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
plt.rcParams.update({'font.size': 18})
```

```python
sum_prog= np.loadtxt("output_ekin.dat")
sum_prog2= np.loadtxt("output_epot.dat")
sum_prog3= np.loadtxt("output_etot.dat")
sum_prog4= np.loadtxt("output_temp.dat")
```

## Esercizio 4.1

In questo primo esercizio ho modificato il codice di dinamica molecolare per ottenere l'equilibrio termo dinamico. Nei grafici sottostanti mostro l'andamento della temperatura e della energia cinetica in funzione del numero di blocchi, la curva si assesta al valore desiderato dalla simulazione

```python
dim=len(sum_prog)
x = np.arange(dim)
plt.figure(figsize=(20,10));
plt.plot(x,sum_prog4[:dim],label="T")
plt.plot(x,sum_prog[:dim],label="Ekin")
plt.plot(x,sum_prog2[:dim],label="Epot")
plt.plot(x,sum_prog3[:dim],label="Etot")
plt.hlines(1.9,0,dim,color='blue')
plt.hlines(2.85, 0, dim,color='orange')
plt.hlines(-4.47559, 0,dim,color='green')
plt.hlines(-1.62559, 0, dim,color='red')
plt.xlabel('#step')
plt.title("Equilibrazione ")
plt.legend()
plt.grid(True)
plt.show()
```

## Esercizio 4.2

### Stato Solido

In questo esercizio mostro la media a blocchi delle varie grandezze fisica del sistema sempre nello stato liquido

```python
ave_kin= np.loadtxt("Solid_result/ave_ekin.out")
ave_pot= np.loadtxt("Solid_result/ave_epot.out")
ave_tot= np.loadtxt("Solid_result/ave_etot.out")
ave_temp= np.loadtxt("Solid_result/ave_temp.out")
```

```python
x = np.arange(len(ave_kin))
fig, ax = plt.subplots(2,2,figsize=(15, 10))

ax[0,0].errorbar(x,ave_kin[:,2],yerr=ave_kin[:,3])
ax[0,1].errorbar(x,ave_temp[:,2],yerr=ave_temp[:,3])
ax[1,0].errorbar(x,ave_tot[:,2],yerr=ave_tot[:,3])
ax[1,1].errorbar(x,ave_pot[:,2],yerr=ave_pot[:,3])

fig.suptitle('Stato Solido')
ax[0,0].set_title('Energia Cinetica')
ax[0,1].set_title('Temperatura')
ax[1,0].set_title('Energia Totale')
ax[1,1].set_title('Energia Potenziale')

plt.show
```

Nelle figure l'asse delle asciesse rappresenta il numero di blocchi su cui  stata esesugita la media. Mentre l'asse delle ordinate rappresenta l'energia in unit di Lennard Jones.
