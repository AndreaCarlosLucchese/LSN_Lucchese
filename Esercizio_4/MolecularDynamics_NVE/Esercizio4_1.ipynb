## Esercizio 4.1

In questo primo esercizio ho modificato il codice di dinamica molecolare per ottenere l'equilibrio termo dinamico. Nei grafici sottostanti mostro l'andamento della temperatura e della energia cinetica in funzione del numero di blocchi, la curva si assesta al valore desiderato dalla simulazione

```python
x = np.arange(1000)
plt.figure(figsize=(20,10));
plt.plot(x,sum_prog4,label="T")
plt.plot(x,sum_prog,label="Ekin")
plt.plot(x,sum_prog2,label="Epot")
plt.plot(x,sum_prog3,label="Etot")
plt.xlabel('#step')
plt.ylabel('Energia')
plt.title("Stato liquido")
plt.legend()
plt.grid(True)
plt.show()
```

## Esercizio 4.2

In questo esercizio mostro la media a blocchi delle varie grandezze fisica del sistema sempre nello stato liquido

```python
ave_kin= np.loadtxt("ave_ekin.out")
ave_pot= np.loadtxt("ave_epot.out")
ave_tot= np.loadtxt("ave_etot.out")
ave_temp= np.loadtxt("ave_temp.out")
```

```python
x = np.arange(100)
fig, ax = plt.subplots(2,2,figsize=(15, 10))

ax[0,0].errorbar(x,ave_kin[:,0],yerr=ave_kin[:,1])
ax[0,1].errorbar(x,ave_pot[:,0],yerr=ave_pot[:,1])
ax[1,0].errorbar(x,ave_tot[:,0],yerr=ave_tot[:,1])
ax[1,1].errorbar(x,ave_temp[:,0],yerr=ave_temp[:,1])

fig.suptitle('Stato liquido')
ax[0,0].set_title('Energia Cinetica')
ax[0,1].set_title('Energia Potenziale')
ax[1,0].set_title('Energia Totale')
ax[1,1].set_title('Temperatura')

plt.show
```

Nelle figure l'asse delle asciesse rappresenta il numero di blocchi su cui è stata esesugita la media. Mentre l'asse delle ordinate rappresenta l'energia in unità di Lennard Jones.
