```python
import tensorflow as tf
from tensorflow import keras
import os
import numpy as np
import matplotlib.pyplot as plt
seed=0
np.random.seed(seed) # fix random seed
tf.random.set_seed(seed)
```

```python
from tensorflow.keras.datasets import mnist
# input image dimensions
# input image dimensions
img_rows, img_cols = 28, 28 # number of pixels 
# output
num_classes = 10 # 10 digits

# the data, split between train and test sets
(X_train, Y_train), (X_test, Y_test) = mnist.load_data()

print('X_train shape:', X_train.shape)
print('Y_train shape:', Y_train.shape)
print()
print(X_train.shape[0], 'train samples')
```

```python
# reshape data, it could depend on Keras backend
X_train = X_train.reshape(X_train.shape[0], img_rows*img_cols)
X_test = X_test.reshape(X_test.shape[0], img_rows*img_cols)

# cast floats to single precesion
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')

# rescale data in interval [0,1]
X_train /= 255
X_test /= 255

# look at an example of data point
print('an example of a data point with label', Y_train[20])
# matshow: display a matrix in a new figure window
plt.matshow(X_train[20,:].reshape(28,28),cmap='binary')
plt.show()

print('an example of a data point with label', Y_train[20], 'before to_categorical ...')
# convert class vectors to binary class matrices, e.g. for use with categorical_crossentropy
Y_train = keras.utils.to_categorical(Y_train, num_classes)
Y_test = keras.utils.to_categorical(Y_test, num_classes)
print('... and with label', Y_train[20], 'after to_categorical')
print()
print('X_train shape:', X_train.shape)
print('Y_train shape:', Y_train.shape)
```

```python
from tensorflow.keras.models import Sequential
from keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import SGD, Adam, RMSprop, Adagrad, Adadelta, Adam, Adamax, Nadam

def create_DNN():
    # instantiate model
    model = Sequential()
    # add a dense all-to-all relu layer
    model.add(Dense(400,input_shape=(img_rows*img_cols,), activation='relu'))
    # add a dense all-to-all relu layer
    model.add(Dense(100, activation='relu'))
    # apply dropout with rate 0.5
   # model.add(Dropout(0.5))
    # soft-max layer
    model.add(Dense(num_classes, activation='softmax'))
    
    return model


def compile_model(appo):
    # create the model
    model=create_DNN()
    # compile the model
    model.compile(loss=keras.losses.categorical_crossentropy,
                  optimizer=appo(),
                  metrics=['accuracy'])
    return model
```

In questo esercizio confronto 3 diversi ottimizzatori Adam, Adagrad e RMS.

```python
# training parameters
batch_size = 32
epochs = 5

# create the deep neural net
model_DNN_Adagrad = compile_model(Adagrad)
model_DNN_Adam= compile_model(Adam)
model_DNN_RMS = compile_model(RMSprop)
# train DNN and store training info in history
history_Adagrad = model_DNN_Adagrad.fit(X_train, Y_train,
          batch_size=batch_size,
          epochs=epochs,
          verbose=1,
          validation_data=(X_test, Y_test))
history_Adam= model_DNN_Adam.fit(X_train, Y_train,
          batch_size=batch_size,
          epochs=epochs,
          verbose=1,
          validation_data=(X_test, Y_test))
history_RMS = model_DNN.fit(X_train, Y_train,
          batch_size=batch_size,
          epochs=epochs,
          verbose=1,
          validation_data=(X_test, Y_test))
```

```python
# look into training history

# summarize history for accuracy
fig,ax = plt.subplots(1,3, figsize=(20, 6))
ax[0].plot(history_Adagrad.history['accuracy'],label="Accuracy")
ax[0].plot(history_Adagrad.history['val_accuracy'],label="Validation Accuracy")
ax[0].set_title("Adagrad")
ax[0].legend()


ax[1].plot(history_Adam.history['accuracy'],label="Accuracy")
ax[1].plot(history_Adam.history['val_accuracy'],label="Validation Accuracy")
ax[1].set_title("Adam")
ax[1].legend()

ax[2].plot(history_RMS.history['accuracy'],label="Accuracy")
ax[2].plot(history_RMS.history['val_accuracy'],label="Validaiton Accuracy")
ax[2].set_title("RMS")
ax[2].legend()


```

Nella mia rete neurale ho deciso di togliere lo strato di dropout poichè peggiorava sensibilmente il training. Possiamo notare come in termini di accuratezza del modello l'ottimizzatore "Adam" risulta essere migliore. Curioso il caso di "Adagrad" in cui la rete ha ottenuto score di accuretezza migliore nella validazione rispetto che nel training.

```python
fig,ax = plt.subplots(1,3, figsize=(20, 6))
ax[0].plot(history_Adagrad.history['loss'],label="Loss")
ax[0].plot(history_Adagrad.history['val_loss'],label="Validation loss")
ax[0].set_title("Adagrad")
ax[0].legend()


ax[1].plot(history_Adam.history['loss'],label="Loss")
ax[1].plot(history_Adam.history['val_loss'],label="Validation loss")
ax[1].set_title("Adam")
ax[1].legend()

ax[2].plot(history_RMS.history['loss'],label="Loss")
ax[2].plot(history_RMS.history['val_loss'],label="Validaiton loss")
ax[2].set_title("RMS")
ax[2].legend()


```

Anche guardando la loss l'ottimizzatore migliore risulta essere Adam tra i 3 che ho preso in considerazione
