```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
plt.rcParams.update({'font.size': 14})
```

```python
sum_prog= np.loadtxt("Econofisica.txt")
sum_prog2= np.loadtxt("Econofisica_2.txt")
```

In questo esercizio mostro il calcolo del prezzo di un opzione call e di un opzione put, con l'utilizzo di due metodi diversi. Il primo metodo, *Valutazione esatta*, consiste nel calcolare il prezzo dell'opzione al tempo esatto in cui l'opzione viene esercitata. Il secondo metodo, *Valutazione con cammino discreto*, consiste invece nel valutare l'opzione diverse volte prima di arrivare al momento in cui viene esercitata l'opzione.

```python
x = np.arange(100)
plt.figure(figsize=(20,10));
plt.errorbar(x,sum_prog[:,0],yerr=sum_prog[:,1],label="Valutazione esatta")
plt.errorbar(x,sum_prog2[:,0],yerr=sum_prog2[:,1],label="Valutazione con cammino discreto ")
plt.xlabel('#Blocchi')
plt.ylabel('Prezzo')
plt.title("Opzione Call")
plt.legend()
plt.grid(True)
plt.show()
```

```python
x = np.arange(100)
plt.figure(figsize=(20,10));
plt.errorbar(x,sum_prog[:,2],yerr=sum_prog[:,3],label="Valutazione esatta")
plt.errorbar(x,sum_prog2[:,2],yerr=sum_prog2[:,3],label="Valutazione con cammino discreto ")
plt.xlabel('#Blocchi')
plt.ylabel('Prezzo')
plt.title("Opzione put")
plt.legend()
plt.grid(True)
plt.show()
```
