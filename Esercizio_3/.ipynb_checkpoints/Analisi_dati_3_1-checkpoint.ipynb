# ANALISI DATI ESERCIZIO 3.1

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
sum_prog= np.loadtxt("Econofisica.txt")
sum_prog2= np.loadtxt("Econofisica_2.txt")
```

```python
x = np.arange(100)
plt.errorbar(x,sum_prog[:,0],yerr=sum_prog[:,1])
plt.errorbar(x,sum_prog2[:,0],yerr=sum_prog2[:,1])
plt.xlabel('#throws')
plt.ylabel('C')
plt.grid(True)
plt.show()
```

```python
x = np.arange(100)
plt.errorbar(x,sum_prog[:,2],yerr=sum_prog[:,3])
plt.errorbar(x,sum_prog2[:,2],yerr=sum_prog2[:,3])
plt.xlabel('#throws')
plt.ylabel('P')
plt.grid(True)
plt.show()
```
